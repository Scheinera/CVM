```
url_dfp = 'https://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/DFP/DADOS/'

from zipfile import ZipFile
arquivos_zip = []
for ano in range(2022,2023):
  arquivos_zip.append(f'dfp_cia_aberta_{ano}.zip')

print(arquivos_zip)

import shutil
shutil.rmtree('/work/Deepnote/CVM/DFP/ZIP')
shutil.rmtree('/work/Deepnote/CVM/DFP/CSV')
shutil.rmtree('/work/Deepnote/CVM/DFP/CONCAT')
shutil.rmtree('/work/Deepnote/CVM/DFP/REPORT')

import os
os.mkdir('/work/Deepnote/CVM/DFP/ZIP')
os.mkdir('/work/Deepnote/CVM/DFP/CSV')
os.mkdir('/work/Deepnote/CVM/DFP/CONCAT')
os.mkdir('/work/Deepnote/CVM/DFP/REPORT')

import wget
for arq in arquivos_zip:
  wget.download(url_dfp+arq, out='/work/Deepnote/CVM/DFP/ZIP')
```

```
from zipfile import ZipFile
import glob

zip_files = glob.glob('/work/Deepnote/CVM/DFP/ZIP/*.zip')

for zip_file in zip_files:
    with ZipFile(zip_file, 'r') as zf:
        zf.extractall('/work/Deepnote/CVM/DFP/CSV')
```

```
import pandas as pd
nomes = ['BPA_con', 'BPA_ind', 'BPP_con', 'BPP_ind', 'DFC_MD_con', 'DFC_MD_ind', 'DFC_MI_con', 'DFC_MI_ind', 'DMPL_con', 'DMPL_ind', 'DRA_con', 'DRA_ind', 'DRE_con', 'DRE_ind', 'DVA_con', 'DVA_ind']
for nome in nomes:
  arquivo = pd.DataFrame()
  for ano in range(2022,2023):
    arquivo = pd.concat([arquivo, pd.read_csv(f'/work/Deepnote/CVM/DFP/CSV/dfp_cia_aberta_{nome}_{ano}.csv', sep=';', decimal=',', encoding='ISO-8859-1')])
  arquivo.to_csv(f'/work/Deepnote/CVM/DFP/CONCAT/dfp_cia_aberta_{nome}_2023', index=False)
```

## Balanço Patrimonial Ativo/Passivo

```
### ATIVO ####

dfp_bpacon = pd.read_csv('/work/Deepnote/CVM/DFP/CONCAT/dfp_cia_aberta_BPA_con_2023')
dfp_bpaind = pd.read_csv('/work/Deepnote/CVM/DFP/CONCAT/dfp_cia_aberta_BPA_ind_2023')

dfp_bpa = pd.concat([dfp_bpacon,dfp_bpaind])
dfp_bpa.to_csv(f'/work/Deepnote/CVM/DFP/REPORT/dfp_cia_aberta_BPA_2023', index=False)

dfp_bpa = pd.read_csv('/work/Deepnote/CVM/DFP/REPORT/dfp_cia_aberta_BPA_2023')
dfp_bpa = dfp_bpa[dfp_bpa['ORDEM_EXERC'] == 'ÚLTIMO']
dfp_bpa = dfp_bpa[dfp_bpa['GRUPO_DFP'] == 'DF Consolidado - Balanço Patrimonial Ativo']

### PASSIVO ####

dfp_bppcon = pd.read_csv('/work/Deepnote/CVM/DFP/CONCAT/dfp_cia_aberta_BPP_con_2023')
dfp_bppind = pd.read_csv('/work/Deepnote/CVM/DFP/CONCAT/dfp_cia_aberta_BPP_ind_2023')

dfp_bpp = pd.concat([dfp_bppcon,dfp_bppind])
dfp_bpp.to_csv(f'/work/Deepnote/CVM/DFP/REPORT/dfp_cia_aberta_BPP_2023', index=False)

dfp_bpp = pd.read_csv('/work/Deepnote/CVM/DFP/REPORT/dfp_cia_aberta_BPP_2023')
dfp_bpp = dfp_bpp[dfp_bpp['ORDEM_EXERC'] == 'ÚLTIMO']
dfp_bpp = dfp_bpp[dfp_bpp['GRUPO_DFP'] == 'DF Consolidado - Balanço Patrimonial Passivo']


### BALANÇO PATRIMONIAL ####

dfp_bp = pd.concat([dfp_bpa,dfp_bpp])

##dfp_bp = dfp_bp[dfp_bp['CD_CVM'] == 1023]
dfp_bp = dfp_bp [['CD_CONTA', 'DS_CONTA', 'VL_CONTA', 'CD_CVM','DENOM_CIA', 'DT_FIM_EXERC']].drop_duplicates().set_index('CD_CONTA')

dfp_bp
```

```
emp_bp = dfp_bp[['DENOM_CIA','CD_CVM']].drop_duplicates().set_index('CD_CVM')
emp_bp
```
