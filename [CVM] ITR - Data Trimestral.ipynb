```
url_itr = 'https://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/ITR/DADOS/'

from zipfile import ZipFile
arquivos_zip = []
for ano in range(2023,2024):
  arquivos_zip.append(f'itr_cia_aberta_{ano}.zip')

print(arquivos_zip)

import shutil
shutil.rmtree('/work/Deepnote/CVM/ITR/ZIP')
shutil.rmtree('/work/Deepnote/CVM/ITR/CSV')
shutil.rmtree('/work/Deepnote/CVM/ITR/CONCAT')
shutil.rmtree('/work/Deepnote/CVM/ITR/REPORT')

import os
os.mkdir('/work/Deepnote/CVM/ITR/ZIP')
os.mkdir('/work/Deepnote/CVM/ITR/CSV')
os.mkdir('/work/Deepnote/CVM/ITR/CONCAT')
os.mkdir('/work/Deepnote/CVM/ITR/REPORT')

import wget
for arq in arquivos_zip:
  wget.download(url_itr+arq, out='/work/Deepnote/CVM/ITR/ZIP/')
```

```
from zipfile import ZipFile
import glob

zip_files = glob.glob('/work/Deepnote/CVM/ITR/ZIP/*.zip')

for zip_file in zip_files:
    with ZipFile(zip_file, 'r') as zf:
        zf.extractall('/work/Deepnote/CVM/ITR/CSV')
```

```
import pandas as pd
nomes = ['BPA_con', 'BPA_ind', 'BPP_con', 'BPP_ind', 'DFC_MD_con', 'DFC_MD_ind', 'DFC_MI_con', 'DFC_MI_ind', 'DMPL_con', 'DMPL_ind', 'DRA_con', 'DRA_ind', 'DRE_con', 'DRE_ind', 'DVA_con', 'DVA_ind']
for nome in nomes:
  arquivo = pd.DataFrame()
  for ano in range(2023,2024):
    arquivo = pd.concat([arquivo, pd.read_csv(f'/work/Deepnote/CVM/ITR/CSV/itr_cia_aberta_{nome}_{ano}.csv', sep=';', decimal=',', encoding='ISO-8859-1')])
  arquivo.to_csv(f'/work/Deepnote/CVM/ITR/CONCAT/itr_cia_aberta_{nome}_2023', index=False)
```

## Balanço Patrimonial Ativo / Passivo - Trimestral

```
### ATIVO ###

itr_bpacon = pd.read_csv('/work/Deepnote/CVM/ITR/CONCAT/itr_cia_aberta_BPA_con_2023')
itr_bpaind = pd.read_csv('/work/Deepnote/CVM/ITR/CONCAT/itr_cia_aberta_BPA_ind_2023')

itr_bpa = pd.concat([itr_bpacon,itr_bpaind])
itr_bpa.to_csv(f'/work/Deepnote/CVM/ITR/REPORT/itr_cia_aberta_BPA_2023', index=False)

itr_bpa = pd.read_csv('/work/Deepnote/CVM/ITR/REPORT/itr_cia_aberta_BPA_2023')
itr_bpa = itr_bpa[itr_bpa['ORDEM_EXERC'] == 'ÚLTIMO']
itr_bpa = itr_bpa[itr_bpa['GRUPO_DFP'] == 'DF Consolidado - Balanço Patrimonial Ativo']

### PASSIVO ####

itr_bppcon = pd.read_csv('/work/Deepnote/CVM/ITR/CONCAT/itr_cia_aberta_BPP_con_2023')
itr_bppind = pd.read_csv('/work/Deepnote/CVM/ITR/CONCAT/itr_cia_aberta_BPP_ind_2023')

itr_bpp = pd.concat([itr_bppcon, itr_bppind])
itr_bpp.to_csv(f'/work/Deepnote/CVM/ITR/REPORT/itr_cia_aberta_BPP_2023', index=False)

itr_bpp = pd.read_csv('/work/Deepnote/CVM/ITR/REPORT/itr_cia_aberta_BPP_2023')
itr_bpp = itr_bpp[itr_bpp['ORDEM_EXERC'] == 'ÚLTIMO']
itr_bpp = itr_bpp[itr_bpp['GRUPO_DFP'] == 'DF Consolidado - Balanço Patrimonial Passivo']


### BALANÇO PATRIMONIAL ####

itr_bp = pd.concat([itr_bpa,itr_bpp])

##itr_bp = itr_bp[itr_bp['CD_CVM'] == 1023]
itr_bp = itr_bp [['CD_CONTA', 'DS_CONTA', 'VL_CONTA', 'CD_CVM','DENOM_CIA', 'DT_FIM_EXERC']].drop_duplicates().set_index('CD_CONTA')

itr_bp
```
